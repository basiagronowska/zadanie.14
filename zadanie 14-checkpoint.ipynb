# Analiza danych

Amplituda - w przebiegach sinusoidalnych jest maksymalną wartością tego przebiegu w innych przypadkach jest to największe wychylenie wartości cech

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = 3
A = 1
t = np.linspace(0, 1, 256)
sygnal = A*np.sin(2*np.pi*f*t)
 
plt.plot(t, sygnal)
plt.show()
```
